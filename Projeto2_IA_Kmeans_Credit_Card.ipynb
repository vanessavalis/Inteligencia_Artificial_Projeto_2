## * Importações das bibliotecas *Plotly* - para geração de gráficos, *Numpy* - para manipulação de matrizes e cálculos matemáticos e *Scikit Learn* - para pré processamento de dados.

```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
from sklearn.preprocessing import StandardScaler
```

## * Importação do algoritmo *KMeans* a partir da biblioteca *Scikit Learn*.

```python
from sklearn.cluster import KMeans
```

## * Importação da biblioteca *Pandas* para fazer a leitura da base de dados .xls.

```python
import pandas as pd
base_cartao = pd.read_excel('/content/credit_card_clients.xls', header = 1)
base_cartao
```

## * Selecionando atributos da base de dados.

```python
X_cartao_atributos = base_cartao.iloc[:, [1,12,13,14,15,16,17]].values
X_cartao_atributos
```

## * Padronização de dados da base utilizando o cálculo da média e do desvio padrão.

```python
scaler_cartao_atributos = StandardScaler()
X_cartao_atributos = scaler_cartao_atributos.fit_transform(X_cartao_atributos)
```

## * Exibição de dados padronizados.

```python
X_cartao_atributos
```

## * Utilização do "método do cotovelo" para definição do número de clusters ideal ao problema.

```python
wcss = []
for i in range (1,11):
  kmeans_cartao_atributos = KMeans(n_clusters = i, random_state = 0)
  kmeans_cartao_atributos.fit(X_cartao_atributos)
  wcss.append(kmeans_cartao_atributos.inertia_)
```

```python
grafico = px.line(x = range(1,11), y = wcss)
grafico.show()
```

```python
kmeans_cartao_atributos = KMeans (n_clusters = 4, random_state = 0)
rotulos = kmeans_cartao_atributos.fit_predict(X_cartao_atributos)
```

```python
rotulos
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
X_cartao_atributos_pca = pca.fit_transform(X_cartao_atributos)
```

```python
X_cartao_atributos_pca.shape
```

```python
X_cartao_atributos_pca
```

```python
grafico = px.scatter(x = X_cartao_atributos_pca[:,0], y = X_cartao_atributos_pca[:,1], color = rotulos)
grafico.show()
```

```python
lista_clientes = np.column_stack((base_cartao, rotulos, X_cartao_atributos_pca))
lista_clientes = lista_clientes[lista_clientes[:, 25].argsort()]
lista_clientes
```

```python
import pandas as pd

tabela_clientes = pd.DataFrame(data = lista_clientes, columns=['ID', 'LIMITE', 'GÊNERO', 'EDUCAÇÃO',  'EST. CIVIL',  'IDADE',  'N. PAG. ATR. 1',  'N. PAG. ATR. 2',  'N. PAG. ATR. 3',  'N. PAG. ATR. 4',  'N. PAG. ATR. 5',  'N. PAG. ATR. 6',  'FATURA 1',  'FATURA 2',  'FATURA 3',  'FATURA 4',  'FATURA 5',  'FATURA 6',  'PAGAMENTO 1',  'PAGAMENTO 2',  'PAGAMENTO 3',  'PAGAMENTO 4',  'PAGAMENTO 5',  'PAGAMENTO 6', 'PAG. PROX. MES', 'CLASSIFICAÇÃO', 'X', 'Y'])
tabela_clientes.head()
```
